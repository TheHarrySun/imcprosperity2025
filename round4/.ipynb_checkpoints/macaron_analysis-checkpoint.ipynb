```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_df(day):
    file_name = f"./round-4-island-data-bottle/observations_round_4_day_{day}.csv"
    return pd.read_csv(file_name, sep=',')
```

```python
df1 = get_df(1)
df2 = get_df(2)
df3 = get_df(3)
df2['timestamp'] += 1000000
df3['timestamp'] += 2000000

df = pd.concat([df1, df2, df3])
```

```python
df.columns
```

```python
df['mid_price'] = (df['bidPrice'] + df['askPrice']) / 2
```

```python
df = df.dropna(subset=['mid_price', 'sugarPrice', 'sunlightIndex', 'timestamp'])
```

```python
csi_range = np.linspace(df['sunlightIndex'].min(), df['sunlightIndex'].max(), 100)
correlations = []
counts = []
```

```python
for csi in csi_range:
    sub_df = df[df['sunlightIndex'] < csi]
    counts.append(len(sub_df))

    if len(sub_df) > 50:
        corr = sub_df['mid_price'].corr(sub_df['sugarPrice'])
    else:
        corr = np.nan
    correlations.append(corr)
```

```python
correlations = np.array(correlations)
best_idx = np.nanargmax(np.abs(correlations))
best_csi = csi_range[best_idx]
best_corr = correlations[best_idx]
print(f"Best CSI: {best_csi:.2f} with correlation: {best_corr:.2f}")
```

```python
plt.figure(figsize = (10,5))
plt.plot(csi_range, correlations, label='Correlation')
plt.axvline(best_csi, color = 'red', linestyle='--', label=f'Best CSI: {best_csi:.2f}')
plt.xlabel("Candidate CSI Threshold")
plt.ylabel("Correlation btwn Mid Price and Sugar Price")
plt.title("Finding the CSI")
plt.legend()
plt.grid(True)
plt.show()
# Probably a good CSI would be around 30-32 because that's when it seems the 2nd 
# derivative becomes more positive
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

df = pd.concat([df1, df2, df3])
df['mid_price'] = (df['bidPrice'] + df['askPrice']) / 2
df = df.dropna(subset = ['mid_price', 'bidPrice', 'askPrice', 'transportFees', 'exportTariff', 'importTariff', 'sunlightIndex', 'sugarPrice'])
```

```python
feature_cols = ['transportFees', 'exportTariff', 'importTariff', 'sunlightIndex', 'sugarPrice']

X = df[feature_cols]
y = df['mid_price']

reg = LinearRegression()
reg.fit(X, y)
y_pred = reg.predict(X)
residuals = y - y_pred
mse = mean_squared_error(y, y_pred)
r2 = reg.score(X, y)
```

```python
print("Model Coefficients:")
for name, coef in zip(feature_cols, reg.coef_):
    print(f"{name}: {coef:.4f}")
print(f"\nIntercept: {reg.intercept_:.4f}")
print(f"R2 Score: {r2:.4f}")
print(f"MSE: {mse:.4f}")

plt.figure(figsize=(10, 5))
plt.scatter(y,y_pred, alpha = 0.4, marker='.')
plt.xlabel("Actual Mid Price")
plt.ylabel("Predicted Mid Price")
plt.title("Multivariate Linear Regression - Predicted vs Actual Mid Price")
plt.grid(True)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.hist(residuals, bins = 50, edgecolor='k')
plt.title("Residual Distribution")
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
CSI = 30.0

df_stressed = df[df['sunlightIndex'] < CSI].copy()
X = df_stressed[['sunlightIndex', 'sugarPrice']]
y = df_stressed['mid_price']

model = LinearRegression()
model.fit(X, y)
sunlight_weight = model.coef_[0]
sugar_weight = model.coef_[1]
intercept = model.intercept_
r2 = model.score(X, y)

print(f"Sunlight weight: {sunlight_weight:.4f}")
print(f"Sugar weight: {sugar_weight:.4f}")
print(f"Intercept: {intercept:.4f}")
print(f"R²: {r2:.4f}")
```

```python
df['conv_sell'] = df['bidPrice'] - df['exportTariff'] - df['transportFees']
df['conv_buy'] = df['askPrice'] + df['importTariff'] + df['transportFees']
df['long_profit'] = df['conv_sell'] - df['mid_price']
df['short_profit'] = df['mid_price'] - df['conv_buy']
```

```python
thresholds = np.linspace(0, 3.0, 50)
conversion_stats = []
for t in thresholds:
    long_trades = df[df['long_profit'] > t]
    short_trades = df[df['short_profit'] > t]
    
    total_trades = len(long_trades) + len(short_trades)
    avg_profit = pd.concat([long_trades['long_profit'], short_trades['short_profit']]).mean()
    total_profit = pd.concat([long_trades['long_profit'], short_trades['short_profit']]).sum()

    conversion_stats.append({
        "threshold": t,
        "num_trades": total_trades,
        "avg_profit": avg_profit,
        "total_profit": total_profit
    })

stats_df = pd.DataFrame(conversion_stats)

fig, ax1 = plt.subplots(figsize =(10,5))
ax1.plot(stats_df["threshold"], stats_df["total_profit"], label="Total Conversion Profit", color="blue")
ax1.set_ylabel("Total Profit")
ax1.set_xlabel("Conversion Threshold")

ax2 = ax1.twinx()
ax2.plot(stats_df["threshold"], stats_df["num_trades"], label="Num Conversions", color="orange")
ax2.set_ylabel("Number of Conversion Opportunities")

plt.title("Conversion Threshold Tuning")
fig.legend(loc="upper right")
plt.grid(True)
plt.show()
```

```python
sunlight_weight = 0.9260
sugar_weight = 10.1435
intercept = -1407.3115
csi = 29

df['panic_fair'] = sunlight_weight * df['sunlightIndex'] + sugar_weight * df['sugarPrice'] + intercept
df['conv_fair'] = ((df['bidPrice'] - df['exportTariff'] - df['transportFees']) + (df['askPrice'] + df['importTariff'] + df['transportFees'])) / 2
df['regime'] = np.where(df['sunlightIndex'] < CSI, 'panic', 'normal')
edge_values = np.linspace(0.2, 2.5, 40)
results = []

for edge in edge_values:
    profit = 0
    trades = 0

    for _, row in df.iterrows():
        if row['regime'] == 'panic':
            fv = row['panic_fair']
        else:
            fv = row['conv_fair']

        if row['askPrice'] < fv - edge:
            profit += fv - row['askPrice']
            trades += 1

        if row['bidPrice'] > fv + edge:
            profit += row['bidPrice'] - fv
            trades += 1
    results.append({"edge": edge,"profit": profit,"trades": trades,"profit_per_trade": profit / trades if trades > 0 else 0})

results_df = pd.DataFrame(results)
print("hi")

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 5))
ax1.plot(results_df['edge'], results_df['profit'], label='Total Profit', color='blue')
ax1.set_xlabel("Edge Value")
ax1.set_ylabel("Total Profit", color='blue')

ax2 = ax1.twinx()
ax2.plot(results_df['edge'], results_df['trades'], label='Number of Trades', color='orange')
ax2.set_ylabel("Number of Trades", color='orange')

plt.title("Tuning Trading Edge for Macarons")
fig.legend(loc='upper right')
plt.grid(True)
plt.show()
```
